### Task 2 · Run Local *(100% offline on AMD)* — edit persona/voice, then type (or talk) in the bar below

*Same LLM, but now Piper-TTS speaks on-device — unplug the network and it keeps talking. Flip `USE_VOICE_CHAT = True` to talk to Reachy: offline speech-to-text (faster-whisper) transcribes your voice, so the whole loop — ears, brain, and voice — runs on this box.*

```python
# >>> TRY ME <<<  Task 2 — fully offline (local LLM + Piper voice, no internet).
# Try: turn off Wi-Fi — Reachy won't miss a beat. Edit PERSONA_2, swap PIPER_MODEL
#      for another voice, or set USE_VOICE_CHAT = True to talk instead of type —
#      then run this cell and use the bar below.
PERSONA_2 = (
    "You are a cute desktop robot assistant. Respond with enthusiasm and warmth, "
    "in two or three short sentences. Keep it brief and conversational."
)
PIPER_MODEL = "en-us-blizzard_lessac-medium"  # en-us-blizzard_lessac-medium (clear) / en_US-amy-medium (warm female) / en_US-ryan-medium (deeper male)
USE_VOICE_CHAT = False  # True = add a 🎤 Speak button (offline speech-to-text) / False = type
```

```python
# Run this cell, then type (or click 🎤 Speak if USE_VOICE_CHAT = True) to chat.
# (Edits to the TRY ME cell above apply on your next message.)
chat_bar(
    get_controller=lambda: get_controller_offline(PIPER_MODEL),
    voice_input=lambda: USE_VOICE_CHAT,
    build_payload=lambda user: {
        "model": CHAT_MODEL,
        "prompt": user,
        "system": PERSONA_2,
        "think": False,
        "keep_alive": "30m",
        "options": {"temperature": 0.8, "num_predict": 120},
    },
)
```
